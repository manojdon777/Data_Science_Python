## apply()

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame(np.random.randint(1,20,size =(6,5)),columns =list('ABCDE'))
df
```

```python
def sqr(n):
    return n**2
df.apply(sqr)
```

```python
df.apply(np.square)
```

```python
df.apply(lambda x : x**2)
```

```python
df['A'] = df['A'].apply(lambda x:x**2)
df
```

```python
df['B'] = df['B'].apply(lambda x:x**3)
df['C'] = df['C'].apply(lambda x:x+100)
df['D'] = df['D'].apply(lambda x:x-100)
df
```

```python
df = pd.DataFrame({'Country':['India','USA','UK'],
                  'Capitals':['Delhi','Washinton','London']})
df
```

```python
def upper(s):
    return s.upper()

df['Capitals'].apply(upper)
```

```python
import re
def clean_str(s):
    s =re.sub('[^A-Za-z]','',s)
    return s

df = pd.DataFrame({'Country':['India 1#',' @ USA','$ UK'],
                  'Capitals':['Delhi','Washinton','London'],
                  'Date':['25 Jan 2022','25 Jan 2022','25 Jan 2022']})
df['Country']=df['Country'].apply(clean_str)
df
```
