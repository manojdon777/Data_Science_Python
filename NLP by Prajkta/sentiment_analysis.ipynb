```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.text,df.label,test_size=0.2,random_state=42)
```

# Preprocessing

### Tokenization

```python
def tokenize(data):
    tokens = word_tokenize(data)
    return tokens
```

```python
x_train_token = x_train.apply(tokenize)
x_test_token = x_test.apply(tokenize)
```

### Cleaning

```python
def cleaning(data):
    text = [i for i in data if i.isalpha()]
    return text 
```

```python
x_train_clean = x_train_token.apply(cleaning)
x_test_clean = x_test_token.apply(cleaning)
```

### Normalization

```python

def Normalization(data):
    text = [i.lower() for i in data]
    return text 
```

```python
x_train_lower = x_train_clean.apply(Normalization)
x_test_lower = x_test_clean.apply(Normalization)
```

```python
x_train_lower
```

### Stopwords Removal

```python
stop = stopwords.words('english')
```

```python

def stopword_1(data):
    text = [i for i in data if i not in stop]
    return text 
```

```python
x_train_without_stop = x_train_lower.apply(stopword_1)
x_test_without_stop = x_test_lower.apply(stopword_1)
```

### Stemming and Lemmatization

```python
def Lemmatization(data):
    lemma = WordNetLemmatizer()
    l1 = []
    for i in data :
        text = lemma.lemmatize(i)
        l1.append(text)
    return l1
```

```python
x_train_lemma = x_train_without_stop.apply(Lemmatization)
x_test_lemma = x_test_without_stop.apply(Lemmatization)
```

```python
x_train_lemma
```

# Count_vectorizer

```python
def join_list(data):
    text = " ".join(data)
    return text
```

```python
final_train = x_train_lemma.apply(join_list)
final_test = x_test_lemma.apply(join_list)
```

```python
final_train
```

```python
from sklearn.feature_extraction.text import CountVectorizer,TfidfVectorizer
```

```python
cv = CountVectorizer(lowercase=True,stop_words='english',max_df=0.95,max_features=1000)
```

```python
count_train = cv.fit_transform(final_train)
count_test = cv.transform(final_test)
```

```python
count_train
```

```python
cv.get_stop_words()
```

```python
cv.get_feature_names()
```

```python
df1 = pd.DataFrame(count_train.A,columns=cv.get_feature_names())
```

```python
df1
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
```

```python
mnb = MultinomialNB()
mnb.fit(count_train.A,y_train)
```

```python
pred_mnb = mnb.predict(count_test.A)
```

```python
pred_mnb
```

```python
acc = accuracy_score(y_test,pred_mnb)
acc*100
```
