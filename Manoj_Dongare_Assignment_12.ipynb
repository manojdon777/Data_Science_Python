#### 7. Write a Pandas program to find and replace the missing values in a given DataFrame which do not have any valuable information.
	â—‹ Example:
	â—‹ Missing values: ?, --
	â—‹ Replace those values with NaN

```python
df1 = pd.DataFrame({'A':[1,2,'?',4,5],'B':[6,7,'-','*',10]})
df1
```

```python
df1.replace({'?':np.nan,'-':np.nan,'*':np.nan},inplace=True)
df1
```

#### 9. Write a Pandas program to drop the rows where all elements are missing in a given DataFrame

```python
import numpy as np
import pandas as pd
array = np.random.randint(10,100,size = [10,3])
df = pd.DataFrame(array, columns = list('ABC'))
df['Z'] = pd.Series([1,2,3,4,5,6,7,8])
df.iloc[1] = np.nan
df.iloc[4] = np.nan
df.iloc[7,1] = np.nan
print('Original DataFrame:\n',df)
# print()
# print('Nan values in columns of given DataFrame:\n',df.isna().sum())
df.dropna(how='all')
```

#### 10. Write a Pandas program to replace all the NaN values with Zero's in a column of a DataFrame

```python
import numpy as np
import pandas as pd
array = np.random.randint(10,100,size = [10,3])
df = pd.DataFrame(array, columns = list('ABC'))
df['Z'] = pd.Series([1,2,3,4,5,6,7,8])
df.iloc[1] = np.nan
df.iloc[4] = np.nan
df.iloc[7,1] = np.nan
df.replace(np.nan,0)
```
